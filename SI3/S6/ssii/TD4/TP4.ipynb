```python
from wget import download
import os

files = [ "cat.zip", "dog.zip", "bird.zip" ]

for file in files:
    if not os.path.exists(file):
        download("https://www.i3s.unice.fr/~lingrand/si3/" + file)
```

```python
# unzip

import zipfile
for file in ("cat.zip", "dog.zip", "bird.zip"):
    if not os.path.exists(file[:-4]):
        with zipfile.ZipFile(file, 'r') as zip_ref:
            zip_ref.extractall()

# Imports
import numpy as np
import matplotlib.pyplot as plt
import librosa
import librosa.display
```

```python
from glob import glob

mfcc = {}

for f in glob("**/*.wav", recursive=True):
    sound, sampling_rate = librosa.load(f)
    print(f)
    mfcc[f] = librosa.feature.mfcc(y=sound, sr=sampling_rate, n_mfcc=13)
    
```
